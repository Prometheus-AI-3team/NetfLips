```python
import torch

# 체크포인트 파일 경로 (확인하고 싶은 파일 경로로 변경하세요)
checkpoint_path = "../speech-resynthesis/checkpoints/unit_av_renderer.pt" 

try:
    # CPU로 로드 (GPU가 없어도 확인 가능하게)
    checkpoint = torch.load(checkpoint_path, map_location='cpu')
    
    # 1. 최상위 키(Key) 확인
    print("=== Top Level Keys ===")
    print(checkpoint.keys())
    print("\n")

    checkpoint = checkpoint['audio']
    print("=== Audio model Keys ===")
    print(checkpoint.keys())
    print()
    checkpoint = checkpoint['en']

    # 2. 'generator' 혹은 모델 State Dict 키 내부 확인
    # (보통 'generator', 'model', 'state_dict' 등의 키를 사용함)
    if 'generator' in checkpoint:
        print("=== Generator State Dict Keys (Layer Names) - First 10 ===")
        # 너무 많으니까 처음 10개만 출력
        for key in list(checkpoint['generator'].keys())[:10]:
            print(key)
            
        # 3. 모델의 Weight Shape 확인 (하나만 예시로)
        first_key = list(checkpoint['generator'].keys())[0]
        print(f"\nShape of '{first_key}': {checkpoint['generator'][first_key].shape}")
        
    elif 'state_dict' in checkpoint:
         print("=== State Dict Keys (Layer Names) - First 10 ===")
         for key in list(checkpoint['state_dict'].keys())[:10]:
            print(key)

    else:
        # 키가 없고 바로 State Dict인 경우
        print("=== Keys (Layer Names) - First 10 ===")
        for key in list(checkpoint.keys())[:10]:
            print(key)

except FileNotFoundError:
    print(f"Error: 파일을 찾을 수 없습니다. 경로를 확인해주세요: {checkpoint_path}")
except Exception as e:
    print(f"Error 발생: {e}")
```
